```python
veer2 = pd.read_csv("/content/cleaned_data.csv")
```

```python
veer2.drop(['Sex'], axis=1, inplace=True)
```

```python
X = veer2.drop(['Survived', 'Name', 'Ticket', 'Cabin','Embarked'], axis=1)
Y = veer2['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train, Y_train)
```

```python
 y_pred = log.predict(X_test)
```

```python
Y_test
```

```python
log.score(X_train, Y_train)
```

```python
log.score(X_test, Y_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(Y_test,y_pred))
```

```python
import seaborn as sns
```

```python
sns.heatmap(confusion_matrix(Y_test,y_pred),annot=True)
```

```python
print(classification_report(Y_test,y_pred))
```

```python
 probabilities = log.predict_proba(X_test)
```

```python
probabilities[:10]
```

```python
predictions = (probabilities[:, 1] >= 0.5).astype(int)
```

```python
import numpy as np
```

```python
results = pd.DataFrame({
    'Probability_Not_Survived': np.round(probabilities[:, 0], 3),
    'Probability_Survived': np.round(probabilities[:, 1], 3),
    'Predicted_Class': predictions})
```
